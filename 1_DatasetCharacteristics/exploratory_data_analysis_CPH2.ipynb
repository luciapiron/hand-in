## Correlations

```python
# Example: Plotting a heatmap to show feature correlations

correlation_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f")
plt.show()
```
